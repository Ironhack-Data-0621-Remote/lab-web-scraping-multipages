## Import libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Store the hot-100 songs list from billboard url in a variable

```python
url = "https://www.billboard.com/charts/hot-100"
```

## Download html with a get request

```python
r = requests.get(url)
r.status_code
```

## Parse html (create the 'soup')

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

## Retrieve/extract the desired info

```python
songs = soup.find_all('li', attrs={'class': 'chart-list__element display--flex'})
songs
```

## Get the text

```python
ranking = []
title = []
artists = []
last_week = []
peak_position = []
weeks_on_chart = []

for i,songs in enumerate(songs):
    ranking.append(songs.find('span', attrs={'class': 'chart-element__rank__number'}).get_text(strip=True))
    title.append(songs.find('span', attrs={'class': 'chart-element__information__song text--truncate color--primary'}).get_text(strip=True))
    artists.append(songs.find('span', attrs={'class': 'chart-element__information__artist text--truncate color--secondary'}).get_text(strip=True))
    last_week.append(songs.find('span', attrs={'class': 'chart-element__information__delta__text text--last'}).get_text(strip=True).split(" ", 1)[0])
    peak_position.append(songs.find('span', attrs={'class': 'chart-element__information__delta__text text--peak'}).get_text(strip=True).split(" ", 1)[0])
    weeks_on_chart.append(songs.find('span', attrs={'class': 'chart-element__information__delta__text text--week'}).get_text(strip=True).split(" ", 1)[0])
```

## Build a dataframe

```python
x = min(len(ranking), len(title), len(artists), len(last_week), len(peak_position), len(weeks_on_chart))

dct = {'ranking': ranking[:x], 'title': title[:x], 'artists': artists[:x], 'last_week': last_week[:x], 'peak_position': peak_position[:x], 'weeks_on_chart': weeks_on_chart[:x]}

df = pd.DataFrame.from_dict(dct)
df
```
